Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Load Dataset

```python
df = pd.read_csv("/content/drive/MyDrive/praktikum_ml/Pertemuan-12/Data/data.csv")
df.head()
```

```python
df = df.dropna(axis=1)
```

 Pisahkan Fitur dan Target

```python
X = df.drop(["id", "diagnosis"], axis=1)
y = df["diagnosis"].map({"B": 0, "M": 1})
```

Standarisasi Data

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Terapkan PCA Menggunakan (2 Komponen)

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)


print("PCA berhasil dijalankan âœ…")
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

Visualisasi Hasil PCA

```python
plt.figure(figsize=(8,6))
plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=y,
    alpha=0.7
)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA pada Breast Cancer Dataset")
plt.show()
```

Klasifikasi Menggunakan Data PCA

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.3, random_state=42
)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("\nAkurasi Klasifikasi (2 PCA Components):", accuracy)
```

Perbandigan (Tanpa PCA)

```python
X_train_o, X_test_o, y_train_o, y_test_o = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)

model_original = LogisticRegression(max_iter=1000)
model_original.fit(X_train_o, y_train_o)

y_pred_o = model_original.predict(X_test_o)

accuracy_original = accuracy_score(y_test_o, y_pred_o)
print("Akurasi Tanpa PCA:", accuracy_original)
```
