```python
import pandas as pd

df = pd.read_csv('/content/gdrive/MyDrive/praktikum_ml/praktikum-3/data/socr.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df1 = (
    df[["Height(Inches)", "Weight(Pounds)"]]
    .rename(columns={"Height(Inches)": "tinggi_cm", "Weight(Pounds)": "berat_kg"})
    .assign(
        tinggi_cm=lambda d: d["tinggi_cm"] * 2.54,      # in → cm
        berat_kg=lambda d: d["berat_kg"] * 0.45359237   # lb → kg
    )
    .round({"tinggi_cm": 2, "berat_kg": 2})
    .copy()
)

# Tampilkan hasil
print(df1.head())
```

```python
from sklearn.model_selection import train_test_split

# Pisahkan fitur (X) dan target (y)
X = df1[["tinggi_cm"]]   # Fitur input
y = df1["berat_kg"]      # Target output

# Bagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=7
)
```

```python
from sklearn.linear_model import LinearRegression

# Membuat objek model regresi linier
model = LinearRegression()

# Melatih model menggunakan data latih
model.fit(X_train, y_train)

```

```python
import numpy as np
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Melakukan prediksi menggunakan data uji
y_pred = model.predict(X_test)

# Menghitung metrik evaluasi
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Menampilkan hasil evaluasi
print("Koefisien (kg per cm):", model.coef_[0])
print("Intersep (kg):", model.intercept_)
print("R2 (test):", r2)
print("MAE (kg):", mae)
print("MSE (kg²):", mse)
print("RMSE (kg):", rmse)
```

```python
# Mengambil nilai slope (koefisien) dan intercept
slope = model.coef_[0]
intercept = model.intercept_

# Menampilkan persamaan regresi
print(f"Persamaan: y = {slope:.3f} * x + {intercept:.3f}")
```

```python
import matplotlib.pyplot as plt

# Plot data scatter
plt.figure(figsize=(8, 5))
plt.scatter(X_test, y_test, color="blue", label="Data Aktual")

# Garis regresi
plt.plot(X_test, y_pred, color="red", linewidth=2, label="Prediksi")

# Label sumbu dan judul
plt.xlabel("Tinggi (cm)")
plt.ylabel("Berat (kg)")
plt.title("Regresi Linear: Berat vs Tinggi")

# Menambahkan teks persamaan dan nilai R2
plt.text(
    0.02, 0.98,
    f"y = {slope:.3f} * x + {intercept:.3f}\nR² = {r2:.3f}",
    transform=plt.gca().transAxes,
    va="top"
)

# Tampilkan legenda dan tata letak
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Melakukan prediksi pada data uji
y_pred_test = model.predict(X_test)

# Membuat tabel hasil (tinggi, berat aktual, berat prediksi, dan error)
hasil = pd.DataFrame({
    "Tinggi (cm)": X_test["tinggi_cm"].to_numpy(),
    "Berat aktual (kg)": y_test.to_numpy(),
    "Berat Prediksi (kg)": y_pred_test,
})

# 1) Hitung selisih error (positif = overpredict)
hasil["Selisih error (kg)"] = hasil["Berat Prediksi (kg)"] - hasil["Berat aktual (kg)"]

# 2) Hitung akurasi per baris (100 * (1 - |error| / aktual)), dibatasi 0–100
denom = hasil["Berat aktual (kg)"].replace(0, np.nan)  # antisipasi pembagi nol
hasil["Akurasi (%)"] = (1 - (hasil["Selisih error (kg)"].abs() / denom)).clip(lower=0, upper=1) * 100

# Tampilkan hasil
hasil
```
